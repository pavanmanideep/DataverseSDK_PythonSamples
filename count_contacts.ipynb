# Count Contacts from Dataverse

This notebook demonstrates how to count contacts in Dataverse using the Web API.

```python
import msal
import requests
```

```python
# Config (reuse your existing values; consider moving to environment variables)
client_id = '0e1c58b1-3d9a-4618-8889-6c6505288d3c'
client_secret = 'qlU8Q~dmhKFfdL1ph2YsLK9URbhIPn~qWmfr1ceL'
tenant_id = '97ae7e35-2f87-418b-9432-6733950f3d5c'
resource = 'https://ecellorsdev.crm8.dynamics.com'
authority = f'https://login.microsoftonline.com/{tenant_id}'
```

```python
# Acquire token
app = msal.ConfidentialClientApplication(client_id, authority=authority, client_credential=client_secret)
print('Acquiring token...')
resp = app.acquire_token_for_client(scopes=[f'{resource}/.default'])
if 'access_token' not in resp:
    print('Failed to acquire token:', resp.get('error'), resp.get('error_description'))
    raise SystemExit(1)
access_token = resp['access_token']
```

```python
# Call Dataverse count endpoint
count_url = f"{resource}/api/data/v9.2/contacts/$count"
headers = {
    'Authorization': f'Bearer {access_token}',
    'Accept': 'text/plain'
}

print('Requesting contact count...')
try:
    r = requests.get(count_url, headers=headers, timeout=15)
    if r.status_code == 200:
        # body contains integer count
        print('Contacts count:', r.text.strip())
    else:
        print('Request failed:', r.status_code)
        print(r.text)
except requests.exceptions.RequestException as e:
    print('Request error:', e)
```
