# Dataverse SDK QuickStart

This notebook demonstrates basic CRUD operations using the PowerPlatform Dataverse SDK.

```python
from azure.identity import InteractiveBrowserCredential
from PowerPlatform.Dataverse.client import DataverseClient
```

```python
# Connect to Dataverse
credential = InteractiveBrowserCredential()
client = DataverseClient("https://ecellorsdev.crm8.dynamics.com", credential)
```

```python
# Create a contact
contact_id = client.create("contact", {"firstname": "John", "lastname": "Doe"})[0]
print(f"Created contact with ID: {contact_id}")
```

```python
# Read the contact back
contact = client.get("contact", contact_id, select=["firstname", "lastname"])
print(f"Contact Created: {contact['firstname']} {contact['lastname']}")
```

```python
# Clean up
client.delete("contact", contact_id)
print("Contact deleted.")
```
